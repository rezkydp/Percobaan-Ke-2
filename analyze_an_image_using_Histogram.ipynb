```python
import cv2
import matplotlib.pyplot as plt

# membaca gambar dalam mode warna
image = cv2.imread('mawarhitam.png', cv2.IMREAD_COLOR)

# mendapatkan resolusi gambar dalam piksel
height, width, channels = image.shape
print(f"Resolusi gambar : {width}x{height} piksel")

# konversi warna dari BGR (OpenCV) ke RGB (Matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# menampilkan gambar
plt.figure(figsize=(5,5))
plt.imshow(image_rgb)
plt.title('Gambar Mawar Hitam')
plt.show()
```

```python
# importing required libraries of opencv
import cv2

# importing library for plotting
from matplotlib import pyplot as plt

# reads an input image
img = cv2.imread('mawarhitam.png',0)

# find frequency of pixels in range 0-255
histr = cv2.calcHist([img],[0],None,[256],[0,256])

# show the plotting graph of an image
plt.plot(histr)
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('mawarhitam.png',0)

# alternative way to find histogram of an image
plt.hist(img.ravel(),256,[0,256])
plt.show()
```
