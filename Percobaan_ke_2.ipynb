```python
# load image
image_1 = cv2.imread('gambar_1.png') # cv2.IMREAD_GRAYSCALE
image_2 = cv2.imread('gambar_2.png')

# Dapatkan resolusi gambar dalam piksel
width, height,channel = image_1.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")

# Dapatkan resolusi gambar dalam piksel
width_2, height_2,channel_2 = image_2.shape
print(f"Resolusi gambar 2: {width_2}x{height_2}x{channel_2} piksel")
```

```python
# plot figure
plt.figure(figsize=(8,4))

plt.subplot(1, 2, 1)
plt.imshow(image_1)
plt.title("Gambar satu")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image_2)
plt.title("Gambar dua")
plt.axis('off')

plt.show()
```

```python
image_1=cv2.resize(image_1,(523,523))
image_2=cv2.resize(image_2,(523,523))
```

```python
# hasil gambar Addition
result_logical = cv2.bitwise_xor(image_1, image_2) # Changed variable names to image_1 and image_2
plt.imshow(result_logical)
plt.show()


# Dapatkan resolusi gambar dalam piksel
width, height, channel = result_logical.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

```python
# hasil gambar Addition
result_logical = cv2.bitwise_and(image_1, image_2) # Changed variable names to image_1 and image_2
plt.imshow(result_logical)
plt.show()


# Dapatkan resolusi gambar dalam piksel
width, height, channel = result_logical.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

```python
# hasil gambar Addition
result_logical = cv2.bitwise_or(image_1, image_2) # Changed variable names to image_1 and image_2
plt.imshow(result_logical)
plt.show()


# Dapatkan resolusi gambar dalam piksel
width, height, channel = result_logical.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

```python
# hasil gambar Addition
result_logical = cv2.bitwise_not(image_1, image_2) # Changed variable names to image_1 and image_2
plt.imshow(result_logical)
plt.show()


# Dapatkan resolusi gambar dalam piksel
width, height, channel = result_logical.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```
